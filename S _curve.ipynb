```python
import numpy as np
import matplotlib.pyplot as plt


T1 = np.linspace(3000,5000,200)

T2 = np.linspace(5000,10000,200)

T3 = np.linspace(10000,50000,200)


n1 = 8
n2 = 5.0
n3 = -3.5

p1 = (13-2*n1)/(4*(7-2*n1))
p2 = (13-2*n2)/(4*(7-2*n2))
p3 = (13-2*n3)/(4*(7-2*n3))

C = 3000


Sigma1 = (T1/C)**(1/p1)
Sigma2 = Sigma1[-1] * (T2/T2[0])**(1/p2)
Sigma3 = Sigma2[-1] * (T3/T3[0])**(1/p3)

#Sigma2 = (T2/C)**(1/p2)

#Sigma3 = (T3/C)**(1/p3)
plt.figure(figsize=(7,5))

plt.loglog(Sigma1,T1,lw=2,label='Cold')

plt.loglog(Sigma2,T2,lw=2,label='Unstable')

plt.loglog(Sigma3,T3,lw=2,label='Hot')

plt.xlabel(r'$\Sigma$')

plt.ylabel(r'$T$')

plt.grid(True,which='both')

plt.legend()

plt.show()
```
